```python
import pandas as pd
import numpy as np
import sklearn.preprocessing
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
users = df.drop(['no','date','line','s_no','name'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(users) 
df_scaled = scaler.transform(users)

df_scaled = pd.DataFrame(data = df_scaled)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(df_scaled)


pcscore = pca.transform(df_scaled)
```

```python
pcscore
```

```python
pca.explained_variance_
```

```python
pca = PCA()
components = pca.fit_transform(df_scaled)
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}
```

```python
pca = PCA(n_components=3)

printcipalComponents = pca.fit_transform(df_scaled)

principalDf = pd.DataFrame(data=printcipalComponents, columns = ['principal component1', 'principal component2', '3'])

```

```python
pca.explained_variance_ratio_
```

```python
pca = PCA(n_components=2) 
printcipalComponents = pca.fit_transform(df_scaled)
principalDf = pd.DataFrame(data=printcipalComponents, columns = ['pc1', 'pc2'])
```

```python
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import collections
import mglearn
import matplotlib.pyplot as plt
import random
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Dendograms")
dend = shc.dendrogram(shc.linkage(principalDf, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit_predict(principalDf)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(principalDf['pc1'], principalDf['pc2'], c=cluster.labels_, cmap='rainbow')
```

```python
from sklearn.cluster import KMeans


ks = range(1,10)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(principalDf)
    inertias.append(model.inertia_)

# Plot ks vs inertias
plt.figure(figsize=(4, 4))

plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('y')
plt.xticks(ks)
plt.show()
```

```python
clust_model = KMeans(n_clusters = 5, 
n_init=10 ,
max_iter=500 ,
random_state = 42,
algorithm='auto'
                    )

clust_model.fit(principalDf) 


centers = clust_model.cluster_centers_ 
pred = clust_model.predict(principalDf) 

print(pd.DataFrame(centers))
print(pred[:10])
```

```python
clust_df = principalDf.copy()
clust_df['clust'] = pred
clust_df.head()
```

```python
plt.figure(figsize=(20, 6))

X = clust_df

plt.subplot(131)
sns.scatterplot(x=X.iloc[:,0], y=X.iloc[:,1], data=principalDf, hue=clust_model.labels_, palette='coolwarm')
plt.scatter(centers[:,0], centers[:,2], c='black', alpha=0.8, s=150)


plt.show()
```
